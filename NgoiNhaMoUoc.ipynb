```python
import os
os.system('cls')
#################
import turtle
t=turtle.Turtle()

turtle.bgcolor("cyan")

def DatConRua(ConRua,x,y):
  ConRua.hideturtle()
  ConRua.penup()
  ConRua.goto(x,y)
  ConRua.pendown()
  ConRua.showturtle()

def HinhChuNhat(ConRua,ngang,doc,MauTo="white"):
  if MauTo=="black":
    MauNet="white"
  else:
    MauNet="black"
  ConRua.color(MauNet,MauTo)
  ConRua.begin_fill()
  ConRua.forward(ngang)
  ConRua.left(90)
  ConRua.forward(doc)
  ConRua.left(90)
  ConRua.forward(ngang)
  ConRua.left(90)
  ConRua.forward(doc)
  ConRua.end_fill()
  ConRua.left(90)
  ConRua.color("black")

def HinhBinhHanh(ConRua,day,ben,MauTo="white",Goc1=15):
  if MauTo=="black":
    MauNet="white"
  else:
    MauNet="black"
  ConRua.color(MauNet,MauTo)
  ConRua.begin_fill()
  ConRua.forward(day)#cạnh đáy
  ConRua.left(Goc1)
  ConRua.forward(ben)#cạnh bên
  ConRua.left(180-Goc1)
  ConRua.forward(day)
  ConRua.left(Goc1)
  ConRua.forward(ben)
  ConRua.end_fill()
  ConRua.left(180-Goc1)
  ConRua.color("black")

def TamGiacVuongCan(ConRua,CanhDay,MauTo):
  if MauTo=="black":
    MauNet="white"
  else:
    MauNet="black"
  ConRua.color(MauNet,MauTo)
  ConRua.begin_fill()
  ConRua.forward(CanhDay)
  ConRua.left(135)
  ConRua.forward(CanhDay*(2**(0.5))*(0.5))
  ConRua.left(90)
  ConRua.forward(CanhDay*(2**(0.5))*(0.5))
  ConRua.end_fill()
  ConRua.left(135)
  ConRua.color("black")


#vẽ mặt trước nhà
DatConRua(t,-250,-205)
HinhChuNhat(t,170,250,"blue")

#Vẽ cửa chính
DatConRua(t,-200,-200)
HinhChuNhat(t,70,150,"#90EE90")#green light

#Vẽ mái nhà
DatConRua(t,-250,45)
TamGiacVuongCan(t,170,"magenta")#đỏ tím

#Vẽ mái nhà
DatConRua(t,-80,45)
t.left(15) #                          Tạo độ nghiêng
HinhBinhHanh(t,170,170*(2**(0.5))*(0.5),"orange",120)

#Vẽ mặt bên
DatConRua(t,-80,-205)
# t.left(15)
HinhBinhHanh(t,170,250,"yellow",75)

#Vẽ cửa sổ
DatConRua(t,-20,-55)
HinhBinhHanh(t,50,70,"brown",75)
t.right(15) #                          Trả độ nghiêng

#Vẽ thân cây
DatConRua(t,200,-205)
HinhChuNhat(t,30,80,"brown")

#Vẽ tán dưới
DatConRua(t,130,-125)
TamGiacVuongCan(t,170,"#90EE90")#green light

#Vẽ tán giữa
DatConRua(t,140,-40)
TamGiacVuongCan(t,150,"#90EE90")#green light

#Vẽ tán trên
DatConRua(t,150,35)
TamGiacVuongCan(t,130,"#90EE90")#green light

#Vẽ mặt trời
DatConRua(t,215,135)
t.color("black","yellow")
t.begin_fill()
t.circle(50)
t.end_fill()
t.color("black")

t.hideturtle()
turtle.done()
```
